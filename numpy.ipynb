## NumPy (Numerical Python) ##
## NumPy is like a super-powered calculator. It handles arrays (grids of numbers) and does math very fast.

## ML often involves huge datasets and mathematical operations — NumPy makes them efficient.

## Think of NumPy arrays as Excel sheets but faster and programmable.
## Quick Answer: NumPy is a Python library for fast numerical computing. Its main feature is the NumPy array, which is like a super‑powered list that can store numbers and perform math operations on entire collections at once. It’s used in data science, machine learning, and scientific computing because it makes handling large datasets and mathematical calculations efficient

```python
import numpy as np

```

```python

# Create an array
scores = np.array([85, 90, 78, 92])

# Do math on the array
print(scores.mean())   # Average score
print(scores.max())    # Highest score
print(scores + 5)      # Add 5 to every score
```

```python
import pandas as pd

# Load a CSV file (student scores)
data = pd.read_csv("C:/Users/User/Desktop/Python Data Analysis/python Analysis.csv")

# Explore the data
print(data.head())       # First 5 rows
print(data.describe())   # Summary statistics
print(data["Math"].mean())  # Average math score
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Simple plot
scores = [85, 90, 78, 92]
plt.plot(scores)
plt.title("Student Scores")
plt.show()

# Seaborn for nicer visuals
sns.histplot(data["Math"], bins=5, kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

```python
import numpy as np

# Convert Math scores to NumPy array
math_scores = np.array(data["Math"])

print("Highest Math Score:", math_scores.max())
print("Lowest Math Score:", math_scores.min())
print("Math Scores + 5:", math_scores + 5)  # Add 5 bonus points
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Line plot of scores
plt.plot(data["Name"], data["Math"], label="Math")
plt.plot(data["Name"], data["English"], label="English")
plt.plot(data["Name"], data["Science"], label="Science")
plt.legend()
plt.title("Student Scores by Subject")
plt.show()

# Histogram of Math scores
sns.histplot(data["Math"], bins=5, kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

```python
# Normal Python list
x = [1, 2, 3, 4]
print(x * 2)   # Output: [1, 2, 3, 4, 1, 2, 3, 4] (repeats list)

# NumPy array
import numpy as np
y = np.array([1, 2, 3, 4])
print(y * 2)   # Output: [2, 4, 6, 8] (multiplies each element)
```

Possible Causes of the Error
1. Column names are misspelled or missing
Make sure "Math", "English", and "Science" are spelled exactly as they appear in your DataFrame. Even a small typo or extra space will cause a KeyError.

✅ Check with:

python
print(data.columns)
2. Missing values or wrong data types
If any of those columns contain non-numeric values (like strings or NaNs), .mean() might fail.

✅ Check with:

python
print(data[["Math", "English", "Science"]].dtypes)
3. "Class" column is missing or misnamed
If "Class" doesn’t exist or is spelled differently (e.g., "class" lowercase), groupby("Class") will fail.

✅ Check with:

python
print(data.columns)
✅ How to Safely Run It
Here’s a safe version that checks everything first:

python
# Check column names
print("Columns:", data.columns)

# Group and calculate mean
try:
    result = data.groupby("Class")[["Math", "English", "Science"]].mean()
    print(result)
except KeyError as e:
    print("KeyError:", e)
