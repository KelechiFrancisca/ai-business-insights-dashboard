```python
CamelCase (sometimes called the “camel method”) is a naming style where words are joined together without spaces, and each word after the first starts with a capital letter. For example:

studentName

totalScore

getUserData

This style is widely used in programming for naming variables, functions, or methods because it makes names easier to read

### What CamelCase Means###
Lower camelCase: The first word starts lowercase, and each following word starts uppercase. Example: myVariableName

Upper CamelCase (PascalCase): Every word starts uppercase. Example: MyVariableName
```

## Quick Answer: np.array is the NumPy function used to create an array — a fast, efficient data structure for storing and manipulating numbers. It turns Python lists (or other sequences) into NumPy arrays, which support powerful mathematical operations across entire datasets at once.

##  What np.array Does
Creates a NumPy array from a Python list, tuple, or other sequence.

Arrays are homogeneous: all elements share the same data type (e.g., integers, floats).

Arrays can be multi‑dimensional: 1D (like a list), 2D (like a matrix), or higher.

Unlike Python lists, arrays allow vectorized operations — math applied to all elements at once.

```python
x = [1, 2, 3, 4]
print(x * 2)   # Output: [1, 2, 3, 4, 1, 2, 3, 4] (repeats list)

# NumPy array
import numpy as np
y = np.array([1, 2, 3, 4])
print(y * 2)   # Output: [2, 4, 6, 8] (multiplies each element)
```

```python
# Visualizing the scores of Student
based_score = [20,40,60,70,80,100]
Chioma = [70,80,90,95,30,89]
David = [40,60,50,45,70,55]

plt.plot(based_score, Chioma,color='red', marker='o', label='Chioma')
plt.plot(based_score, David, marker='x',color='purple', label='David')

plt.title("A Graph of Students score")
plt.ylabel("Main score")
plt.xlabel('Students Scores')

plt.legend()

plt.close
```
